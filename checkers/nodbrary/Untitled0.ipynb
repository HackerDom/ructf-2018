```python
import numpy as np
import pandas as pd
from faker import Faker

N = 1000
```

```python
df = pd.read_csv('books.csv')
df.columns
```

```python
re = "^[a-zA-Z0-9,.()\-\"!?;:\'= ]*$"
df = df[["original_title", "authors", "original_publication_year"]].dropna().reset_index(drop=True)
df = df[df["original_title"].str.contains(re) & df["authors"].str.contains(re)]
df.head(10)
```

```python
faker = Faker()
df['publisher'] = pd.Series([faker.company() for _ in range(2*N)])
df['description'] = pd.Series([faker.text(30).replace("\n", " ") for _ in range(2*N)])
df['content'] = pd.Series([faker.text(1000).replace("\n", " ") for _ in range(2*N)])
df['original_publication_year'] = df['original_publication_year'].astype(int)
df = df.rename(columns={"original_publication_year":"year", "original_title":"name", "authors":"author"})
df = df.head(N)
df.head()
```

```python
df.to_csv('result.csv', index=False)
```
